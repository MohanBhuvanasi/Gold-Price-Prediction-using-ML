importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics

```

data collection and processing

```python
#loading the csv data to a pandas dataframe
gold_data= pd.read_csv('/content/gld_price_data.csv')
```

```python
#print first 5 rows in the data frame
gold_data.head()
```

```python
#print the last five rows of the data frame
gold_data.tail()
```

```python
#getting some basic info about data
gold_data.info()
```

```python
#checking the no of missing values
gold_data.isnull().sum()
```

```python
from IPython.utils.text import DollarFormatter
# getting the statistical measures of the data
gold_data.describe()
```

correlation:
1.positive, 2. negative

```python
correlation = gold_data.corr(numeric_only=True)
correlation
```

```python
#constructing a head map to understand the correlation
plt.figure(figsize=(8,8))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
#correlation values of Gld
print(correlation['GLD'])
```

```python
#check the distribution of the gold price
sns.distplot(gold_data['GLD'],color='green')
```

splitting the features and target

```python
x=gold_data.drop(['Date','GLD'],axis=1)
y=gold_data['GLD']
```

```python
print(x)
```

```python
print(y)
```

splitting into training data and test data

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

model training:random forest regressor


```python
regressor=RandomForestRegressor(n_estimators=100)
```

```python
#"training the model"
regressor.fit(X_train,Y_train)
```

model evaluation

```python
#prediction on test data
test_data_prediction=regressor.predict(X_test)
```

```python
print(test_data_prediction)
```

```python
#R square error
error_score=metrics.r2_score(Y_test,test_data_prediction)
print("R squared error:",error_score)
```

compare the actual values and predicted values in polot

```python
Y_test=list(Y_test)
```

```python
plt.plot(Y_test,color='blue',label='Actual Value')
plt.plot(test_data_prediction,color='green',label='Predicted Value')
plt.title('Actual Price vs Predicted Price')
plt.xlabel('Number od values')
plt.ylabel('GLD Price')
plt.legend()
plt.show()
```
